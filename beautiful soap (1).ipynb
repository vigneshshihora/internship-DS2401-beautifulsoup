```python
import requests
from bs4 import BeautifulSoup

url = "https://www.cnbc.com/world/?region=world"


response = requests.get(url)


soup = BeautifulSoup(response.content, "html.parser")


headings = soup.find_all("div", class_="MarketsBanner-teaserContainer")
dates = soup.find_all("span", class_="QuoteStrip-lastTradeTime")
links = soup.find_all("a", class_="QuickLinks-quickLink")


headings_text = [heading.text.strip() for heading in headings]
dates_text = [date.text.strip() for date in dates]
links_href = [link['href'] for link in links]


for i in range(len(headings_text)):
    print("Heading:", headings_text[i])
    print("Date:", dates_text[i])
    print("Link:", links_href[i])
    print()
```

```python
posts = soup.find_all('div', class_='sc-dlVxhl gJYudr')

for post in posts:
    heading = post.find('h2', class_='sc-dlVxhl gJYudr__title').text.strip()
    date = post.find('time', class_='sc-dlVxhl gJYudr__date').text.strip()
    content = post.find('div', class_='sc-dlVxhl gJYudr__content').text.strip()

  
    youtube_link = post.find('a', class_='sc-dlVxhl gJYudr__embed__video')
    if youtube_link:
        youtube_url = youtube_link['href']
       
        video_id = youtube_url.split('/')[-1]
        
        print("Heading:", heading)
        print("Date:", date)
        print("Content:", content)
        print("YouTube Video URL:", youtube_url)
        print()

    else:
        print("Heading:", heading)
        print("Date:", date)
        print("Content:", content)
        print("No YouTube Video found.")
        print()
```

```python
def scrape_product_details():
   

    products = soup.find_all('div', class_='brand-name  undefined')

    for product in products[:10]:
       
        name = product.find('h3', class_='Men's Black Batman Outline Logo T-shirt').text.strip()

        
        price = product.find('span', class_='discountedPriceText clr-p-black   false  ').text.strip()

       
        image_url = product.find('img', class_='productCardImg')['src']

        print("Product Name:", name)
        print("Price:", price)
        print("Image URL:", image_url)
        print('-' * 50)

if __name__ == "__main__":
    print("Scraping product details from Bewakoof.com:")
    scrape_product_details()
```
